```python
original_text=""" Health Policy and Technology 10 (2021) 209–229

Contents lists available at ScienceDirect

Health Policy and Technology
journal homepage: www.elsevier.com/locate/hlpt

Literature Review

Telemedicine and health policy: A systematic review
Clemens Scott Kruse∗, Kelly Williams1, John Bohls1, Waleed Shamsi1
School of Health Administration, Texas State University, San Marcos, TX, United States

a r t i c l e

i n f o

Article history:
Available online 29 October 2020
Keywords:
Health
Policy
Telemedicine
Access
Health policy

a b s t r a c t
Background: Telemedicine diagnoses and treats patients remotely via telecommunications technology all
over the world. Telemedicine becomes more prevalent as providers recognize the beneﬁts, patients receive increased access and payers see the reduction in cost of care.
Objective: Telemedicine studies have shown success in limiting geographical constraints, time spent, and
costs incurred by patients with positive health outcomes across medical specialties. The aim of this review is to evaluate the implications of telemedicine and health policies.
Methods: An assessment of the literature in four databases was made on content germane to health
policy implications of telemedicine. From the results of the search, 48 publications were kept for analysis.
Results: The ﬁfteen facilitators mentioned most often were increased access, increased convenience, improved population health, care enabled through mobile technology, self-eﬃcacy, increased patient-toprovider communication, cost advantages, eﬃcacy of modality, increased health outcomes, reaches developing countries, increased quality, a positive previous experience, and a secure means of care. The
twelve barriers mentioned most often were the increased cost to providers, patient privacy, technical literacy, state licensing, data security, socioeconomics, limited reimbursements, issues of interoperability,
patient safety, less personal means of care, misaligned incentives, and ethical concerns.
Conclusions: Telemedicine has the potential for growth and adoption, however, there are several implications and barriers of health policy surrounding telemedicine that make it diﬃcult to adopt. Policies will
likely encourage and incentivize its spread and use. Future research should focus on standardization of
telemedicine and new policies and incentives that encourage its use.
© 2020 Fellowship of Postgraduate Medicine. Published by Elsevier Ltd.
This is an open access article under the CC BY-NC-ND license
(http://creativecommons.org/licenses/by-nc-nd/4.0/)

Introduction
Rationale
Health policy and telemedicine are heavily intertwined just as
any innovation needs to be regulated for eﬃcacy and safety, payment models need to be established and implemented, and privacy
and security need to be maintained [1–3]. With telemedicine and
new legislative health policies, outcomes can be improved with
greater eﬃciency for both patient and provider [2–4]. This can be

Abbreviations: ACA, affordable care act; CINAHL, cumulative index of nursing
and allied health literature; HIT, health information technology; HITECH, health
information technology for economic and clinical health; MeSH, medical subject
headings; PRISMA, preferred reporting items for systematic reviews and metaanalyses; WHO, World Health Organization; WoS, web of science.
∗
Corresponding author at: School of Health Administration, Texas State University, 601 University Dr, Encino Hall, Room 250, San Marcos, TX 78666, United States.
E-mail address: scottkruse@txstate.edu (C.S. Kruse).
1
These authors contributed equally.

done by reaching patients in a more comfortable setting, removing
travel and time constraints, while improving the patient’s health
and experience. Telemedicine has been creeping into the healthcare ﬁeld for decades, but diffusion of this innovation will likely
improve with the implementation of additional health policy such
as the Affordable Care Act (ACA) and the Health Information Technology for Economic and Clinical Health (HITECH) Act in America which expanded organizational models that use telemedicine
and broadened reimbursement mechanisms for the same [5,6]. It
is logical that providers would examine all options of eﬃcacious
care and gravitate to those that are cost effective, but health policy must enable the innovation, establish basic standards of safety
and security, and reduce boundaries to their adoption.
Telemedicine alters the modality of delivery for healthcare.
Care is still provided in a high-quality manner, but it is delivered
through telecommunications technology [7]. This modality becomes easier and less costly with the spread of technology, speciﬁcally smartphones [8]. Telemedicine is the solution to the problem
of proximate distance between providers and patients with mobil-

https://doi.org/10.1016/j.hlpt.2020.10.006
2211-8837/© 2020 Fellowship of Postgraduate Medicine. Published by Elsevier Ltd. This is an open access article under the CC BY-NC-ND license
(http://creativecommons.org/licenses/by-nc-nd/4.0/)

C.S. Kruse, K. Williams, J. Bohls et al.

Health Policy and Technology 10 (2021) 209–229

ity restrictions. It is vital that the patients seeking care have access
to broadband internet to fully take advantage of the full services of
providers. The spread of telemedicine is reliant on health policy as
it would allow and encourage the use of telemedicine to improve
the Iron Triangle of cost, quality and access.
Health policies that implement and enforce speciﬁc policies
such as cross-state interoperability and reimbursement models will
greatly inﬂuence and increase the reach of telemedicine [3,9–11].
The HITECH Act serves as a key example of health policy playing
an integral role in the spread of technology in healthcare. This was
enacted in 2009 to incentivize and later penalize those who did
not currently use electronic health records [6].

Study selection
Our team independently screened all abstracts in pairs, and we
conducted a consensus meeting to discuss disagreement [12]. We
calculated a Kappa score to provide a measure of agreement between reviewers.

Data collection and data items
We used a standardized Excel spreadsheet as a data extraction tool, in accordance with the protocol [12]. This tool acted as
template for reviewers to collect study design, participants, sample
size, intervention, observed bias, and effect size, where applicable.
The spreadsheet became a literature matrix to list and organize all
articles, extract data between multiple reviewers, and discuss observations in consensus meetings. Three consensus meetings were
held for reviewers to discuss disagreement and share observations.
This practice created a synergistic effect and ensured everyone progressed with a like mind.

Objective
This review examines the interaction between telemedicine and
health policy by identifying facilitators and barriers to the adoption
of telemedicine that health policy could promote and overcome,
respectively. Articles and studies that report on this interaction, regardless of study design, number or types of participants, will be
analyzed through narrative analysis for themes.
Methods

Risk of bias in individual studies and across studies

Protocol and registration

As reviewers read each article, they made note of any observation of bias. For example, when an article makes conclusions
across a population, but 75% of the sample used was female, selection bias would be noted because the external validity of the
study is limited due to the sample. The Effective Public Health
Practice Project (EPHPP) tool was used as a quality assessment
tool for quantitative studies [14]. For the other studies and articles,
we used Johns Hopkins Nursing Evidence-Based Practice (JNHEBP)
tool [15]. This tool collects data on studies and formulates a score
on six components, one of which is selection bias. Much of the
data for this tool were already collected with the Excel spreadsheet
mentioned above, so determination of bias was not diﬃcult. We
listed our observations of bias, such as selection bias and surveillance bias, on the spreadsheet for analysis and summary.

We used the published protocol for systematic reviews in a
health-related academic program for conducting this review [12].
No human subjects were involved with this review, and it was not
registered.
Eligibility criteria
Articles were eligible for this review if they were published in
English in the last ﬁve years. Study design, types and number of
participants did not matter. The key acceptance criterion was if the
article reported on the intersection of telemedicine and health policy. Of particular interest were indicators of barriers or facilitators
to adoption. Articles that did not address barriers, facilitators, or
health policy implications were rejected.
Information sources

Synthesis of results

Authors conducted a systematic review of the literature by
gathering data from PubMed (MEDLINE), the Cumulative Index of
Nursing and Allied Health Literature (CINAHL), Embase, and Web
of Science databases. Authors conducted searches between 15 and
16 June 2020.

We used the Excel spreadsheet to synthesize our observations
and the data collected. The spreadsheet enabled a narrative analysis which identiﬁed themes, as is the practice in multiple disciplines [16]. We did not combine results of studies because very
few studies used similar statistics and reported effect sizes.

Search
Additional analysis
Reviwers identiﬁed keywords from our initial research and
matched those with the terms on Medical Subject Headings
(MeSH). The terms used to conduct this search were health policy,
policy, health, telehealth, mobile health, mHealth, telehealth, and
eHealth. We chose this combination of terms due to the deﬁnition
of telemedicine from the World Health Organization: “. . . the use
of information communications technology to improve patient outcomes by increasing access to care and medical information” [13].
We used the following inclusion criteria in database ﬁlters, when
the database enabled them: full text, English, and the publication
dates of January 2015- June 2020. We excluded reviews. We excluded MEDLINE records, where possible, to eliminate duplicates.
We used the same Boolean phrase for all databases: ("health policy" OR "Policy, Health") AND ("telehealth" OR "mobile hHealth"
OR "health, mobile" OR mhealth OR telehealth OR ehealth).

We performed a thematic analysis on the articles analyzed to
identify patterns in the observations [17]. Not all observations
could be combined into a theme, so we reported them as single
observations. Reviewers read each article two times [12]. The ﬁrst
reading was to make observations of ﬁndings on the Excel data
extraction tool. These observations were reviewed and discussed
to identify trends during the second consensus meeting. During
the second reading, reviewers made independent notes of major
themes related to the objective, using the Excel data extraction
tool. After a third consensus meeting debrieﬁng the observations
and themes, detailed notes were formulated about health policy
implications of telemedicine. Frequency of occurrence of each of
the major, common themes were captured in aﬃnity matrices for
further analysis. Data and calculations are available upon request.
210

C.S. Kruse, K. Williams, J. Bohls et al.

Health Policy and Technology 10 (2021) 209–229

Fig. 1. Literature search and selection process.

Results

title, study design, participants, sample size, intervention, bias,
and observations about barriers or facilitators to the adoption of
telemedicine. We identiﬁed implications of health policy through
analysis of these 48 articles. A full listing of the logic to the thematic analysis is provided in Appendix B.

Study selection and study characteristics
The study selection process is illustrated in Fig. 1. Search terms
were entered into PubMed, Web of Science, Embase (Science Direct), and CINAHL. The authors narrowed the publication date in
both searches to 2015-2020 to ensure the most recent data for
health policy implications of telemedicine. We chose a ﬁve-year
time span because telemedicine is a highly advancing ﬁeld and
ﬁve years provided ample data to analyze. After applying ﬁlters to
PubMed, Web of Science, Embase, and CINAHL 225 remained. As
described in the Methods section, we screened abstracts applicability to the objective statement. When the abstract did not provide
suﬃcient information, we read the full study to make the determination of discard or keep for analysis. We narrowed the list of
germane studies to 48 for analysis.
The results from the search did not always provide fruitful articles that met the screening criteria. For instance, one article listed
telemedicine as an extension of nursing education: Telemedicine
was not used in any clinical capacity, but rather it was used as a
remote meeting tool with preceptors. This use is outside the scope
of how telemedicine is deﬁned, so this article was discarded. Another article listed telemedicine for use in nursing school clinicals
with simulated patients. This, again, is outside the scope because
simulated patients cannot discuss their diﬃculties or experiences
using the system, report medical outcomes, and the article did not
discuss any public policy implications.
We calculated a Kappa score of .65 which is indicative of moderate agreement [18,19]. The calculation of the Kappa score can be
seen in Fig. 2.
The kappa statistic was calculated with a series of Chi Squares
(see Fig. 3) [19]. A full version of the calculations is provided in
Appendix A.
The 48 articles were read and analyzed by at least 2 reviewers [12]. The details extracted were: Year of publication, authors,

Risk of bias within studies
Bias was not detected in all articles. Most of the bias observed
was a restriction of participants to a local population, which limits the external validity of the results. There was one convenience
sample [20] and one instance of surveillance bias [9].
Results of individual studies
General observations and thematic analysis are listed in Table 1.
Articles are listed in reverse 9,20-28chronological order: 2020
(2) [28,29], 2019 (9) [9,20–24,25–27], 2018 (20) [1,4,5,8,10,30–43],
2017 (9) [3,11,44–50,2,3,11,46–50], 2016 (6) [51–56], 2015 (4) [57–
60]. A table of PICOS is provided in Appendix C.
The review examined articles that combine telehealth and
health policy or included facilitators and barriers to telehealth. A
physical count of observations was recorded into Excel for analysis.
Once the Excel ﬁle was completed, it was organized and broken
down into common themes. These facilitator and barrier themes
were then displayed in an aﬃnity matrix to establish frequency of
occurrence. The reviewers do not suggest that frequency equates
to importance. The frequency identiﬁes the statistical probability
that the theme would occur in our analysis. Frequency highlights
salient themes to the literature reviewed.
Risk of bias across studies
Appendix A provides additional details about the articles analyzed including bias. Most studies included some bias such
as restricting the study to the country in which the re211

C.S. Kruse, K. Williams, J. Bohls et al.

Health Policy and Technology 10 (2021) 209–229

Fig. 2. Calculation of Kappa score.

Fig. 3. Example of how Kappa statistic is calculated.

searchers lived [1,3–5,8–11,20–24,25,27–30,32–47,49–60,35,36,39–
42,46–48,50]. Included in the analysis were grey literature such as
opinion pieces and protocols, which do not contain samples, results, or bias [3,5,27,30,43,49,51,53,57,1,2,8,31,35,43,61]. Two studies demonstrated some bias selection bias and one reported
surveillance effect [9,20].

Additional analysis
Countries from Which Articles Originated
Table 2 lists the countries from which articles originated. The
largest number of studies originated in the United States (17/48,
35%), followed by Canada (6/48, 13%). Three studies originated each
212

C.S. Kruse, K. Williams, J. Bohls et al.

Health Policy and Technology 10 (2021) 209–229

Table 1
Review of the literature.
Authors

Implication to policy

Observation themes

Timpel P. and
Harst L. [28]

Public funding of research should extend to include patient safety,
eﬃcacy of modality, technical literacy, ethical issues, data security,
and cost.

Aiyegbajeje F. and
Ajayi D. [29]

Due to proliferation of smart mobile devices, telemedicine services
should extend to this modality.
Telemedicine favors the patient in terms of convenience
Due to proliferation of smart mobile devices, telemedicine services
should extend to this modality.
Telemedicine is effective, it encourages self care, and is often
preferred over traditional care.

Patient safety
Eﬃcacy of modality
Technical literacy
Can raise ethical issues
Data security
Increased cost (to provider) because reimbursement does not cover
the technology
Care enabled through smart mobile technology

Alhassan et al. [21]
Zobair et al. [22]

Ehrenreich K. and
Marston C. [22]

Telemedicine favors the patient in terms of convenience

Spears et al. [24]

Telemedicine can be enabled through smart phones, but ownership
of this technology can be dependent on socioeconomics.
Reminder calls could be considered as a standard rather than a best
practice
SMS reminders could be considered a standard rather than a best
practice

Luckmann et al.
[25]
Wagstaff et al. [26]

Vesnic-Alujevic
et al. [27]

Donelan et al. [20]

Morita et al. [9]

Moerenhout et al.
[30]
Alonso-Solis et al.
[31]
Appiah et al. [32]
Bierman et al. [10]

Chabot et al. [4]

CMS [5]
Gonsalves et al.
[33]
Griﬃths et al. [34]

Guo et al. [35]
Katapally [36]

Based on trends in the EU, the integration of data from wearable
sensors into medical records is problematic. The data are not
consistent, have not been scrubbed, are not veriﬁable, and its
transmission is not secure.
Telemedicine favors the patient in terms of convenience

Due to proliferation of smart mobile devices, telemedicine services
should extend to this modality.
Technical literacy will hinder wide adoption.
Telemedicine has been shown to improve population health through
self care
Telemedicine can be a platform for health education.
Due to proliferation of smart mobile devices, telemedicine services
should extend to this modality.
No consistency across states in terms of tehehealth policy, and many
states used different terminology which could complicate a
comprehensive, nation-wide policy in the future.
Although telemedicine can be perceived as less personal, this quality
can be advantageous when a patient does not want the personal
experience.
Telehealth extends care to rural areas where access to care is not as
robust.
Telemedicine policy can integrate adaptive algorithms to show areas
of greatest effect.
There are many advantages to the use of telemedicine, and it
encourages self-care. There are concerns over data security, patient
safety, and the increased cost to the provider for start-up and
maintenance costs.

Due to proliferation of smart mobile devices, telemedicine services
should extend to this modality.
Telemedicine improves some areas of population health.

Increased convenience (time, miles, work, parking, childcare, etc.)
Care enables other resources (educational/technological)
Self-eﬃcacy
User experience using modality
Eﬃcacy of modality
A positive previous experience
Modality is preferred traditional care
Increased convenience (time, miles, work, parking, childcare, etc.)
Care enables other resources (educational/technological)
Increased access
Modality is preferred traditional care
Secure means of care
Increased access
Care enabled through smart mobile technology
Socioeconomics
Eﬃcacy of modality
Care enables other resources (educational/technological)
Care enables other resources (educational/technological)
Reaches developing countries
Increased patient-to-provider communication
Data security
Data quality
Increased cost (to provider) because reimbursement does not cover
the technology
Increased convenience (time, miles, work, parking, childcare, etc.)
Modality is preferred traditional care
Eﬃcacy of modality
Improved population health
Less personal
Increased patient-to-provider communication
Increased cost (to provider) because reimbursement does not cover
the technology
Care enabled through smart mobile technology
A positive previous experience
Eﬃcacy of modality
Technical literacy
Improved population health
Self-eﬃcacy
Care enables other resources (educational/technological)
Care enabled through smart mobile technology
Reaches developing countries
State licensing

Care enables other resources (educational/technological)
Increased access
Modality is preferred traditional care
Increased access
Improved population health
Improved population health
Self-eﬃcacy
Can raise ethical issues
Data security
Patient safety
Increased cost (to provider) because reimbursement does not cover
the technology
Increased patient-to-provider communication
Care enabled through smart mobile technology
Increased access
Improved population health
(continued on next page)

213

C.S. Kruse, K. Williams, J. Bohls et al.

Health Policy and Technology 10 (2021) 209–229

Table 1 (continued)
Authors

Implication to policy

Observation themes

Kjelle et al. [37]

Policy must consider ﬁnancial barriers to adoption.
National policy must include needs of local leadership.
Policy can include cost savings and improved safety of limited
transfers.
Telemedicine favors the patient in terms of convenience

Limited reimbursement
Misaligned incentives
Cost advantages

Kotz et al. [1]

Health policy must integrate data security to protect patient privacy
and security.

Lee et al. [38]

Policy in developing countries can focus on mHealth ﬁrst because a
cellular network is usually present.

LeGrand et al. [8]

Marginalized populations are particularly at risk of being overlooked
by telemedicine policy.
Telemedicine increases access, but there are cost, quality, and
security concerns.

Leung et al. [39]

Melchiorre et al.
[40]

There are many advantages to the use of telemedicine. Inadequate
funding could be compensated by cost eﬃciencies.

Saleh et al. [41]

Health policy must include displaced populations such as refugees.

Watkins et al. [42]

This article discusses the health policy implications of telemedicine
as increased access to healthcare services and the need to enable
educational resources.

Wyatt [43]

Health policy must include quality assessment criteria for health
apps to ensure eﬃcacy of modality.

Rokicki S. and Fink
G. [44]
Razmak J. and
Belanger C. [45]
Grustam et al. [46]

Due to proliferation of smart mobile devices, telemedicine services
should extend to this modality.
Due to proliferation of smart mobile devices, telemedicine services
should extend to this modality.
This article discusses how telemedicine can improve the treatment of
chronic heart failure through mHealth

Hatef et al. [47]
Malvey M. and
Slovensky D. [3]

Pariyo et al. [49]

Sevick et al. [50]

Due to proliferation of smart mobile devices, telemedicine services
should extend to this modality.
Due to proliferation of smart mobile devices, telemedicine services
should extend to this modality.
Policy should verify eﬃcacy of each application but reduce barriers
to approval to encourage innovation.
Due to proliferation of smart mobile devices, telemedicine services
should extend to this modality.
Policy must include evidence-based interventions, performance
management, enhanced communication, and political commitment
Electronic discharge tools can become a best practice

Increased convenience (time, miles, work, parking, childcare, etc.)
Increased cost (to provider) because reimbursement does not cover
the technology
Secure means of care
Data security
Patient privacy
Care enabled through smart mobile technology
Improved population health
Increased access
Improved care for certain diseases
Reaches developing countries
Socioeconomics
Care enables other resources (educational/technological)
Increased patient-to-provider communication
Care enables other resources (educational/technological)
Increased access
Less personal
Data security
Issues of interoperability
Increased cost (to provider) because reimbursement does not cover
the technology
Improved care for certain diseases
Increased patient-to-provider communication
Self-eﬃcacy
Increased convenience (time, miles, work, parking, childcare, etc.)
Cost advantages
Increased cost (to provider) because reimbursement does not cover
the technology
Issues of interoperability
Technical literacy
Patient privacy
Increased access
Care enables other resources (educational/technological)
Increase health outcomes
Increased access
Reaches developing countries
Care enables other resources (educational/technological)
Increased convenience (time, miles, work, parking, childcare, etc.)
Socioeconomics
Increase health outcomes
Increased access
Improved care for certain diseases
Increased access
Patient safety
Self-eﬃcacy
Care enables other resources (educational/technological)
Care enables other resources (educational/technological)
Improved care for certain diseases
Increase quality
Self-eﬃcacy
Care enabled through smart mobile technology
Limited reimbursement
Improved population health
Improved interoperability
Increased access
Misaligned incentives
Can raise ethical issues
Improved population health
Socioeconomics
Reaches developing countries
Increased convenience (time, miles, work, parking, childcare, etc.)
Cost advantages
Increase health outcomes
Increased cost (to provider) because reimbursement does not cover
the technology
(continued on next page)

214

C.S. Kruse, K. Williams, J. Bohls et al.

Health Policy and Technology 10 (2021) 209–229

Table 1 (continued)
Authors

Implication to policy

Observation themes

Zelmer et al. [11]

This article identiﬁes the need to enable technical resources and the
ability to improve interoperability as implications of policies for
telemedicine.

Utidjian L. and
Abramson E. [51]

Telemedicine favors the patient in terms of convenience
85% of acute, ambulatory pediatric encounters could take place
through telemedicine

Issues of interoperability
Increased access
Improved care for certain diseases
Increased convenience (time, miles, work, parking, childcare, etc.)
Increased patient-to-provider communication
Increased convenience (time, miles, work, parking, childcare, etc.)

Martin et al. [52]

Telemedicine policy should include dentristy.

Slater et al. [53]

Telemedicine policy should include musculoskeletal treatments.

Salisbury et al.
[54]

Telemedicine policy should include mental health

Rechel et al. [55]

Due to proliferation of smart mobile devices, telemedicine services
should extend to this modality.
A national policy would help limit variability between states.

Schmeida M. and
McNeal R. [56]

Kwong et al. [57]

Telehealth resource centers should extend their services to enable all
aspects of telemedicine

Shorten et al. [58]

Telemedicine policy should include decision aids.

Knight et al. [59]

Policy should focus on rural health since residents there have the
least access to local care.
Policy should consider start-up and maintenance costs, privacy, and
standards of interoperability.

Mukooyo et al.
[60]

from Ghana, the UK, and Australia (3/48, 6%). Two studies originated each from China and the EU (2/48, 4%). Twelve studies all
originated from a country speciﬁed only once in the group of articles for analysis.
Although a large variety of interventions were used, the most
common was mHealth (8/48, 17%). The next-most common inter-

Cost advantages
Eﬃcacy of modality
A positive previous experience
Increased access
Increase quality
Care enables other resources (educational/technological)
Lack of infrastructure
Integration into workﬂow
Provider resistance
Technical literacy
Patient privacy
Technical literacy
Less personal
Increased cost (to provider) because reimbursement does not cover
the technology
Limited reimbursement
Unclear ROI
Misaligned incentives
Credentialing concerns
State licensing
Liability concerns
Socioeconomics
Care enables other resources (educational/technological)
Cost advantages
Care enabled through smart mobile technology
Care enables other resources (educational/technological)
Increase health outcomes
Increase quality
Increased access
Self-eﬃcacy
Increased patient-to-provider communication
Technical literacy
A positive previous experience
Increase quality
Self-eﬃcacy
Increase health outcomes
State licensing
Increased access
Limited reimbursement
Patient privacy
State licensing
Cost advantages
State licensing
Increased access
Cost advantages
Increase quality
Increase health outcomes
Increased cost (to provider) because reimbursement does not cover
the technology
Patient privacy
Care enables other resources (educational/technological)
Patient privacy
Increased access
A local telephone triage overcomes impersonal nature
Increased cost (to provider) because reimbursement does not cover
the technology
Issues of interoperability
Patient privacy
Increased access

vention was eHealth (5/48, 10%). These two were also mentioned
in combination with other modalities of telemedicine. Other interventions listed were social media, computer models for prediction
algorithms, phone calls, electronic messages (regardless of platform), surveys, and live video.
Facilitators to the Adoption of Telemedicine

215

C.S. Kruse, K. Williams, J. Bohls et al.

Health Policy and Technology 10 (2021) 209–229

Table 2
Countries from which articles originated.
Country

Occurrences (n=48) Probability of occurrence

References

United States
[1,3,5,8,10,20,23,24,25,33,47,49,51,52,56–58] 17
Canada
[4,9,11,36,45,50]
6
3
Ghana
[21,32,44]
UK
[34,43,54]
3
Australia
[22,53,59]
3
China
[35,39]
2
EU
[27,40]
2
Germany
[28]
1
Norway
[37]
1
Belgium
[30]
1
South Africa
[42]
1
[60]
1
Uganda
Nigeria
[29]
1
Spain, Hungary, Israel
[31]
1
United States, Netherlands, and South Africa
[26]
1
Sub-Saharan Africa
[38]
1
Lebanon
[41]
1
Singapore, Netherlands, United States
[46]
1
Europe, United States, Canada, Australia, Italy, Spain, UK, Croatia, Estonia [55]
1

35%
13%
6%
6%
6%
4%
4%
2%
2%
2%
2%
2%
2%
2%
2%
2%
2%
2%
2%

Table 3
Frequencies of facilitators to adoption of telemedicine.
Themes of facilitators to adoption

References

Occurrences (n=116)

Probability of occurrence

Increased access
Care enables other resources (educational/technological)
Increased convenience (time, miles, work, parking, childcare, etc.)
Improved population health
Care enabled through smart mobile technology
Self-eﬃcacy
Increased patient-to-provider communication
Cost advantages
Eﬃcacy of modality
Increase health outcomes
Reaches developing countries
Increase quality
A positive previous experience
Modality is preferred traditional care
Secure means of care
Improved interoperability
A local telephone triage overcomes impersonal nature
User experience using modality

[1,3–5,11,23,35,39–42,51,53,55,57,59,60]
[4,8,21,23,25,26,31,38,41,42,44,45,51–53,58]
[11,20,23,29,37,40,42,50,51]
[1,20,30,33,34,36,47,49]
[1,9,24,29,32,35,46,53]
[22,30,34,40,43,46,53,54]
[8,11,20,26,34,40,53]
[37,40,50,51,53,57]
[9,20,22,25,28,51]
[41,42,50,53,54,57]
[26,32,38,41,49]
[46,51,53,54,57]
[9,22,51,54]
[4,20,22,23]
[1,23]
[3]
[59]
[22]

18
16
9
8
8
8
7
7
6
6
5
5
4
4
2
1
1
1

16%
14%
8%
7%
7%
7%
6%
6%
5%
5%
4%
4%
3%
3%
2%
1%
1%
1%

[1-5,11,36,38,40,42,43,491,4,20,34,35,37,48,50,628,32,39,42,43,493,10,11,4133,39,42,5011,20,38,519,33,368,11,351,4,44].

The reviewers identiﬁed facilitators and barriers to the adoption of telemedicine in the 48 articles with the intent of identifying where health policy could impact adoption. Table 3 lists the
ﬁfteen themes and three individual observations that were identiﬁed as facilitators to the adoption of telemedicine.
The review identiﬁed increased access in 18/116 occurrences
(18%)
[1–5,11,35,37,39,41,42,48,1,3–5,11,23,35,39–
42,51,53,55,57,59,60]. This theme comprised observations such
as without the telemedicine service, the participant would not
have been able to reach this service, and the participant chose
telemedicine for the privacy aspects of the modality. Authors noted
telemedicine’s ability to extend care to vulnerable populations and
rural populations. Researchers used mHealth, eHealth, telephone
triage, and structured interviews for interventions. Six of these
articles originated from the United States, while two originated
from China, Australia, and Canada. Other countries only occurred
once.

The second most mentioned theme was that care enables other
capabilities such as medical education which occurred in 16/116 occurrences (14%) [4,8,21,23,25,26,31,38,41,42,44,45,51–53,58]. Young
adult participants were receptive to receiving education on sexually transmitted infections through their mobile phones. Education
through mobile technology enabled tailored, enhanced decision
making. SMS recipients were more likely to return to the clinic for
treatment. Smart phone users are more likely to seek information
on the Internet which makes them more likely to receive education
through this same device. Education through mobile phones is a
national extension of using the same modality for sharing information through social media. Researchers used mHealth and eHealth
as interventions for this group. Six of these articles originated from
the United States, three from countries in Africa, two from Canada,
and one from Australia.
The theme next most identiﬁed was an increased convenience
on the part of participants, which occurred 9/116 occurrences (8%)

216

C.S. Kruse, K. Williams, J. Bohls et al.

Health Policy and Technology 10 (2021) 209–229

[11,20,23,29,37,40,42,50,51]. Participants noted signiﬁcant improvements to quality of life speciﬁcally in the logistical details surrounding care such as miles driven, waiting time, parking and
childcare challenges, not taking time off work, and social issues
of judgment by others seeing them enter a clinic. Interventions
used were structured questionnaires, virtual video, mHealth, and
eHealth. Only three of these originated from the United States,
two from Canada, countries in Africa, and one from the EU and
Norway.
The theme next most identiﬁed was an improvement in population health for speciﬁc populations, which occurred 8/116 occurrences (7%) [1,20,30,33,34,36,47,49]. This theme sounds like an
outcome, but it was identiﬁed as a theme because it best summarized the observations such as improvement in HIV population, improvement in diabetes population, improvement in dementia population. Interventions used were mHealth, wearable monitors, and
virtual video visits. Five of the seven articles originated from the
United States: The others originated from Belgium, Canada, and the
UK.
The theme next most identiﬁed was care enabled through
smart, mobile technology, which occurred 8/116 occurrences (7%)
[1,9,24,29,32,35,46,53]. Although, conceptually, this theme has
some overlap with the theme of convenience, there is only one article that truly overlaps. The theme was a combination of observations from participants that they appreciated the modality because
it enabled them to avoid taking off work, being near a computer,
or even access health services at the worksite or in rural environments. Participants already own this modality, so for them, it is a
natural extension of their daily lives. Interventions in this theme
were mHealth, structured questionnaires, and wearable monitors.
Two of these articles originated in the United States and two in
countries in Africa. The rest originated from only one country or
one combination of countries.
The theme next most identiﬁed was that telemedicine
enhanced abilities for self-eﬃcacy, which occurred 7/116 occurrences (6%) [22,30,34,40,43,46,53,54]. Participants noted
telemedicine enabled them to understand more about their
disease. It changed participant behavior in a way that affected their condition positively and encouraged self-care.
Interventions were mHealth, eHealth, digital communications, and structured questionnaires. Three studies originated
in the UK, two were from Australia, the others were single
originations.
The theme next most identiﬁed was that telemedicine increased
patient-to-provider communication, which occurred 7/116 occurrences (6%) [8,11,20,26,34,40,53]. Both participants and providers
noted telemedicine enhanced existing patient-to-provider communication because technology had already been inserted into
the medical transaction. Interventions used in these studies
were mHealth, eHealth, and video visits. Two of these studies originated in the United States. The others were from single
originations.
The theme next most identiﬁed was that telemedicine enabled cost advantages, which occurred 7/116 occurrences (6%)
[37,40,50,51,53,57]. Researchers noted cost savings that occur
through telemedicine because the modality supplants transfers to
higher level of care. This cost advantage can often counter the
costs associated with start-up and maintenance of the equipment
necessary for providers to deliver services through telemedicine.
Interventions used in these articles were eHealth and electronic
communication tools. Three of these articles originated in the
United States. The others were from single originations.
The theme next most identiﬁed was the eﬃcacy of the
modality of telemedicine, which occurred 6/116 occurrences (5%)
[9,20,22,25,28,51]. Researchers noted an improved understanding
of the effectiveness of telemedicine for certain diseases, and that

in many instances, the telemedicine intervention was as clinically
effective as face-to-face care. Interventions used were survey instruments, structured questionnaires, phone calls, virtual video visits, and mHealth. Three of these articles originated from the United
States, while the others were from single originations.
The theme next most identiﬁed was the increase in health outcomes, which occurred 6/116 occurrences (5%) [41,42,50,53,54,57].
Researchers noted the telemedicine intervention introduced
shorter, more frequent visits, as well as earlier interventions.
Telemedicine was used in an HIV study and was found to be
particularly effective at changing behavior for healthier outcomes.
Telemedicine was used in a diabetes study and was highly effective
in getting patients to have their annual eye exam. Interventions
used were mHealth, wearable monitors, digital communications,
virtual video visits, and computer models to predict behavior.
Five of these articles originated from the United States, while the
others were from single originations.
The theme next most identiﬁed was that telemedicine reaches
developing countries, which occurred 5/116 occurrences (4%)
[26,32,38,41,49]. Researchers noted the effectiveness of SMS messaging in developing countries because a cellular network is in
place before a robust Internet infrastructure. The only intervention
used for this group of articles was mHealth. Articles were all from
single originations or a single group of originations: Ghana, SubSaharan Africa, Lebanon, United States, and the grouping of United
States, Netherlands, and South Africa.
The theme next most identiﬁed was that telemedicine increases
quality which occurred 5/116 occurrences (4%) [46,51,53,54,57]. Researchers noted that telemonitoring is safe and reliable. They noticed a higher level of data quality and completeness of the medical record when telemedicine interventions were used. Follow-ups
enabled better continuity of care. The telemedicine modality enabled more services with higher quality care, improved safety, and
less error. The only intervention used for this group of articles was
eHealth. Two of these articles originated from the United States,
while the others were from single originations.
The theme next most identiﬁed was the history of a patient’s positive previous experience with telemedicine, which occurred 4/116 occurrences (3%) [9,22,51,54]. Researchers noted that
a positive previous experience with telemedicine predisposed individuals to be willing to try telemedicine for future interventions,
and as a result of positive previous experiences, participants were
more likely to report a positive experience with future interventions. A positive previous experience with telemedicine enabled
high patient satisfaction scores in future encounters. Interventions
used were structured questionnaires, mHealth, and eHealth. All articles were from single originations, but all were developed nations.
The theme next most identiﬁed was modality of telemedicine
was preferred over traditional modalities, which occurred 4/116 occurrences (3%) [4,20,22,23]. Several participants stated that they
preferred telemedicine over traditional, in-person care. Some of
these observations were indicative of the setting: One study concerned abortion access in which the state required a 72 hour wait
period between two appointments. When the ﬁrst appointment
could be performed through telemedicine, participants were interested, which led to their strong preference. Others noted a feeling of enjoyment from using a telemedicine intervention. Other
strongly preferred telemedicine because it avoided a crowded waiting room. Interventions used were questionnaires and virtual video
visits. Three of these articles originated from the United States,
while the fourth was from Canada.
The theme next most identiﬁed was that telemedicine was a secure means of care, which occurred 2/116 occurrences (2%) [1,23].
Researchers noted that in some settings, traditional face-to-face
care can enable guilt: Some participants noted they could not go to
217

C.S. Kruse, K. Williams, J. Bohls et al.

Health Policy and Technology 10 (2021) 209–229

Table 4
Frequencies of barriers to adoption of telemedicine.
Themes of barriers to adoption

References

Occurrences (n=64)

Probability of occurrence

Increased cost (to provider) because reimbursement does
not cover the technology
Patient privacy
Technical literacy
State licensing
Data security
Socioeconomics
Limited reimbursement
Issues of interoperability
Patient safety
Less personal
Misaligned incentives
Can raise ethical issues
Credentialing concerns
Unclear ROI
Liability concerns
Data quality
Integration into workﬂow
Provider resistance
Lack of infrastructure

[20,27,28,34,37,39,40,50,51,57,60]

11

17%

[1,40,51,57,58,60]
[9,28,40,51,53]
[10,51,55–57]
[1,27,28,34,39]
[24,38,42,49,52]
[37,46,51,56]
[11,39,40,60]
[28,34,43]
[20,39,51]
[3,37,51]
[3,28,34]
[51]
[51]
[51]
[27]
[51]
[51]
[51]

7
6
5
5
5
4
4
3
3
3
3
1
1
1
1
1
1
1

9%
9%
8%
8%
8%
6%
5%
5%
5%
5%
5%
2%
2%
2%
2%
2%
2%
2%

a local clinic out of fear of judgment from neighbors. Telemedicine
enabled the privacy of seeking care from their home on a secure connection to prevent eavesdropping. Interventions used were
structured interviews and wearable monitors. Both of these articles
originated from the United States.
The next set were not themes because they only occurred once
in the literature (1%): improved interoperability [3]. a local telephone
triage overcomes an impression of impersonal care [59], and level of
user experience using modality [22]. Improved interoperability was
offered as an advantage to the existence of policy and standards as
technology enters a developing nation because developers would
be forced to develop with interoperability in mind before many solutions are available in the speciﬁc market. This article originated
in the United States and was an opinion article on global mHealth
policy. The other two articles originated from Australia and were
studies that used structured questionnaires and a telephone triage
system.
Barriers to the Adoption of Telemedicine
Twelve themes and seven individual observations were identiﬁed as barriers to the adoption of telemedicine. These are listed in
Table 4 by frequency of occurrence.
The review identiﬁed increased cost in 11/64 occurrences (17%)
[20,27,28,34,37,39,40,50,51,57,60]. Researchers noted an increased
cost to providers due to reimbursements not covering the cost of
the technology to either start delivering or the maintenance costs
associated with continuing to provide telemedicine services. Interventions used were eHealth, mHealth, survey instruments, virtual video visits, wearable sensors, digital communications, and
social-media platforms. Three of these articles originated from
the United States, two were from the EU, while the rest were
from single originations: Uganda, Germany, Norway, Canada, and
the UK.
The next barrier mentioned was patient privacy in 7/64 occurrences (9%) [1,40,51,56–58,60]. Researchers noted the inconsistency
in state health policy regarding telemedicine creates a large diversity of protections for patients. Interventions used were eHealth
and wearable monitors. Five of these articles originated from the
United States, while the others were from the EU and Uganda.
Technical literacy was identiﬁed 6/64 occurrences (9%)
[9,28,40,51,53]. Researchers noted those who do not own a
smart phone may be less technologically savvy than those who

do own one. Providers were concerned their staff possess a lack
of technical skills and a general discomfort with technology, and
these factors would inhibit their ability to offer telemedicine
services. Interventions used were mHealth, eHealth, and a survey instrument. All these articles were from single originations:
Germany, Canada, EU, United States, and Australia.
State licensing was identiﬁed 5/64 occurrences (8%) [10,51,55–
57]. Researchers noted disparities in health policy between states
in the United States, creating a large difference in how occupational therapy and physical therapy is supported. The lack of a national policy makes it diﬃcult for providers to virtually cross state
lines to offer their services. Some states enable the license of certain other states to validate services in their own, but this is not
universal, making things unnecessarily complicated for providers.
Interventions used were survey instruments and video. Three of
the four articles originated from the United States, while the fourth
included the United States, Europe, and Canada.
Data security was also identiﬁed 5/64 occurrences (8%)
[1,27,28,34,39]. Researchers noted concerns about data security,
data protection, and inadvertent disclosure of information. Interventions used were digital communications, survey instruments,
wearable monitoring, and social-media platforms. All articles were
from single originations: Germany, EU, UK, United States, and
China.
Socioeconomics was also identiﬁed in 5/64 occurrences (8%)
[24,38,42,49,52]. Researchers noted the socioeconomics between
owning a smart phone and a regular phone. Airtime was also a
critical factor for those plans that pay per SMS message, because
this would, of course, limit the success of an mHealth intervention that used SMS as its primary means to interface with patients.
Interventions used were all mHealth. Three of these articles originated from the United States while the other two came from countries in Africa.
Limited reimbursement was identiﬁed in 4/64 occurrences (6%)
[37,46,51,56]. This theme was similar to increase in cost, but only
two of the articles overlapped. Researchers noted the limits in
available reimbursements for telemedicine interventions limited
the services that they could offer. The low level of the reimbursement of those that existed did not cover their costs of the
technology necessary to offer the intervention. Interventions used
were mobile services. Two of these articles originated from United
218

C.S. Kruse, K. Williams, J. Bohls et al.

Health Policy and Technology 10 (2021) 209–229

States, while the others came from Norway and a combination of
Singapore and the Netherlands.
We also identiﬁed issues of interoperability in 4/64 occurrences
(6%) [11,39,40,60]. Researchers noted the existence of very few
standards for the interface of sensors, eHealth, and mHealth interventions with the electronic health record. Interventions used
were mHealth, eHealth, and social-media platforms. These articles all came from single originations: China, Europe, Canada, and
Uganda.
We identiﬁed patient safety in 3/64 occurrences (5%) [28,34,43].
Researchers noted the need to vet health applications to verify the
veracity of the data that they collect as well as the physical safety
of their use. Interventions used were mHealth, digital communications, and survey instruments. Two of these articles originated
from the UK and one from Germany.
Less personal was identiﬁed in 3/64 occurrences (5%) [20,39,51].
Researchers noted clinicians felt a loss of personal connection
with patients when using telemedicine interventions, and the lack
of face-to-face interaction could be deleterious to the patient-toprovider relationship. Interventions used virtual video visits and
social-media platforms. Two of these articles originated from the
United States, while the other was from China.
Misaligned incentives was identiﬁed in 3/64 occurrences (5%)
[3,37,51]. Researchers noted the dichotomy of incentives that currently exist. Tele-referrals are easily enabled through telemedicine,
but the referring provider is not incentivized to refer: The provider
is incentivized to keep the patient in house. Referring a patient out
is clearly in the best interest of the patient, but incentives do not
reﬂect that fact. Interventions used were mobile services. Two of
these articles originated from the United States, and the other was
from Norway.
We identiﬁed ethical concerns in 3/64 occurrences (5%)
[3,28,34]. Clinicians are concerned that easy access to health
providers through telemedicine may undermine a feeling of personal responsibility for decisions. Interventions used were survey
instruments and digital communication. These articles came from
single originations: Germany, UK, and the United States.
The next set of observations were not considered themes because they only occurred once in the literature (2%), and six of
them came from the same article: credentialing concerns, unclear
return on investment (ROI), liability concerns, integration into workﬂow, provider resistance, and lack of infrastructure [51] and data
quality [27]. Six of the observations originated from an article in
the United States, and the other originated in the EU.

their performance management, evidence-based design, eﬃcacy,
and safety to users [3,28,49].
Although a majority of the articles in the review originated
from the United States, of the seventeen mHealth interventions,
only ﬁve originated from the United States, four from Canada,
three from African countries, and the others from single originations: China, Lebanon, UK, and two multi-national studies. Six of
the mHealth articles highlighted the medical education extension
the modality can enable [26,31,41,42,44,58]. A large concern in the
United States is the issue of licensing across state lines. Efforts toward a national health policy on telemedicine would alleviate this
concern.
Interestingly, only ﬁve articles used eHealth in the intervention,
and none of these originated from the United States. These articles
had single originations: Uganda, UK, Canada, EU, Australia. Studies
in these countries evaluated telemedicine’s ability to treat arthritis,
mental health, and multiple modalities [40,45,53,54,60].
Not surprisingly, an increase in access was the most frequently mentioned facilitator in the literature. An increase in access refers to the ability for patients to access their providers
at all levels of the healthcare continuum: Primary care, emergency medicine, mental health [1–5,11,35,37,39,41,42,48]. During
the COVID-19 pandemic, telemedicine was relied upon heavily to
extend care where in-person was either not necessary or not feasible [62] For instance, ED providers relied upon telemedicine to
triage forward: prior to arrival to the ED, enabling quarantine without infecting others in the waiting room. Other providers in other
levels of the healthcare continuum also saw patients at a distance.
This serves as an excellent example of how policy can be quickly
passed or adapted to enable care at a distance.
Care through the modality of telemedicine enables other resources such as education. The technology enables care givers to
push educational modules concerning diabetes, wound care, mental health, instructional modules on how to administer medication, and preventative care such as prenatal text messages. In all
cases, participants were willing to accept the education through
their phones, and that education was instrumental in changing behavior to include healthier choices [8,31,38,41,42,48].
Issues of interoperability appeared as both a facilitator and a
barrier. Interoperability enables providers to better communicate
within their community to share medical information with peers
in disparate organizations. As a facilitator, one researcher proposed
establishing policy early in a developing nation to set standards
of interoperability up front before developers move in with solutions. The barrier of interoperability appeared mostly expressing
concern that such a practice did not occur in developed nations.
As a result, existing telemedicine interventions do not easily interface with electronic health records, which complicates documentation and exacerbates an already strained administrative burden
[3,10,11,40].
An important feature of mHealth is that it relies upon a cellular
network, which is often in place before a robust Internet infrastructure. Simple Messaging System (SMS) interventions can even
occur without data service. This simple protocol has shown great
promise in developing nations. Telemedicine enables providers to
reach remote areas of developing countries where the broadband
is not robust, but a cellular network exists. Mhealth primarily uses
text messages or the SMS protocol. This protocol does not require a
data network, but instead transmits small packet messages through
a cellular network [32,38,41,49].
The modality of telemedicine is not free. There are signiﬁcant
up-front costs and some baseline training of both providers and
staff to use the technology in a safe, secure, and effective manner. The ongoing costs are usually negligible, are certainly less
than the step-cost associated with expanding clinic space. The upfront costs to patients are almost negligible, as long as they al-

Discussion
Summary of evidence
In this review, the authors identiﬁed 27 themes and 10 individual observations of facilitators or barriers to adoption of
telemedicine, and each has implications to health policy. This information is relevant to policy makers as they are the ones who
need to create policies that can increase the use of telemedicine
as to provide better quality care to more people. Several articles highlighted the proliferation of cellular technology throughout the world, and because this technology and cellular network already exist, the market is primed for mHealth interventions [3,9,21,29,32,44–46,55]. Articles highlighted characteristics of
telemedicine that favor the patient in terms of convenience and
privacy [20,23,29,51], however reimbursement and incentives do
not align with this modality [3,37,46,51,56]. Another article identiﬁed 85% of acute, ambulatory, pediatric encounters could take
place through telemedicine [51]. Researchers agreed that government should regulate the market of health applications to ensure
219

C.S. Kruse, K. Williams, J. Bohls et al.

Health Policy and Technology 10 (2021) 209–229

ready own the telecommunications technology on which it will be
hosted [20,34,50].
Although some might think ownership of technology is common, there exists a very clear and well-documented technology
gap that is based on race, socioeconomics, and age. People of color
own less technology. Lower socioeconomic sectors of society own
less technology. Older adults do not always own or trust technology [2,9,42]. There is a very good reason for not trusting technology: Technology can be hacked. Personal and ﬁnancial information
can be leaked. Live sessions can be eavesdropped on. Organized
crime feeds off medical information due to its rich opportunities
on the dark web [3,63].
Telemedicine is continuing to increase in prevalence across the
country and the world with its exciting potential to extend the
boundaries of physician practice. It can overcome the barriers of
proximity [11,20,37,50] and reach rural patients and those in developing countries to provide coordinated care [32,38,41,49]. Our
results from this systematic review identiﬁed several facilitators
and barriers to the adoption of telemedicine and their correlation to health policy. Barriers consist of current health policy limiting providers within states, cost of implementation, ethical concerns, resource limitation, the digital age gap and the current reimbursement model [2,3,9,20,34,42,50]. Reimbursement mechanisms
must exist to enable this modality in the healthcare industry of
nations. In the U.S., 50 states establish their own standards for
the practice of telemedicine, and many of these states require the
provider to reside in the state in which telemedicine is being utilized: The situation is similar in Europe [64] Such practices defeat
telemedicine’s purpose to heal at a distance, through the Internet
or mobile device, and in a secure manner [1,4,7,9,32,35,43]. Health
policy makers needs to break down these barriers rather than
erecting new ones. As this new modality of care becomes more
widely accepted and preferred, the health policy of nations will
have to adjust and expand to govern and monitor it while incentivizing providers and patients to use it. Facilitators of telemedicine
include increased access, enabling resources effectively, improving quality and patient experience, improved interoperability, improved population health and reaching developing countries [1–
5,8–11,20,31–43,46–50,61]. The health policy makers of nations
should leverage these facilitators to break down barriers to adoption. Policymakers need to recognize the value of the telemedicine
modality of care and push for legislation that encourages or incentivizes its spread and use.

responsible for article selection. To mitigate selection bias, multiple
reviewers independently screened and analyzed articles. Publication bias is also possible as journals typically publish articles with
notable ﬁndings over those with minimal statistical signiﬁcance.
Publication bias was mitigated using multiple research databases:
e.g., CINAHL also includes grey literature in its search results.
Conclusions
Research supports health policies that increase access, improve
population health, and remove constraints to reach patients who
live in developing countries and rural areas, and align both incentives and reimbursement to support the technology necessary to
deliver these services. Telemedicine has the potential for growth
and adoption. Health policies need to promote this to reduce costs
while increasing access. Through continued evaluation and further adoption, research may ﬁnd how beneﬁcial and affordable
telemedicine can be. As health policies standardize and promote
telemedicine use across state lines, rural patients especially will
see the results of improved population health and patient experience [2,47,49].
Authors’ contribution
CSK provided structure, guidance, critical analysis of 100% of articles, and served as senior editor. KW provided project management of the research, she assigned tasks, analyzed 33% of the articles, wrote the Methods section and created initial ﬁgures. WS
analyzed 33% of the articles and created the literature matrix. JB
served as the editor, analyzed 33% of the articles, and wrote the
introduction, results, and eligibility criteria sections.
Ethical approval
Not required.
Funding
This research received no speciﬁc grant from any funding
agency in the public, commercial, or not-for-proﬁt sectors.
Declaration of Competing Interest
The Author(s) declare(s) that there is no conﬂict of interest.

Limitations
Acknowledgments
Authors noted minor limitations. The high variability between
the articles, telemedicine monitoring methods, patient samples and
treatment types could compromise the external validity of the literature. Selection bias must also be noted as the researchers were

None.

220

C.S. Kruse, K. Williams, J. Bohls et al.

Health Policy and Technology 10 (2021) 209–229

Appendix A. Kappa calculation by light, interpreted by McHugh

221

C.S. Kruse, K. Williams, J. Bohls et al.

Health Policy and Technology 10 (2021) 209–229

Appendix B. thematic analysis

Authors

Observations

Observation Themes

Timpel P. and
Harst L. [28]

Improved understanding of patient safety

Patient safety

Improved understanding of effectiveness
Improved understanding of technology acceptance
Improved understanding of ethical issues.
Improved understanding of data protection
Improved understanding of costs

Eﬃcacy of modality
Technical literacy
Can raise ethical issues
Data security
Increased cost (to provider) because reimbursement does not cover
the technology
Care enabled through smart mobile technology

Aiyegbajeje F. and
Ajayi D [29]

Mobile communication between patients and medical doctors
predicted a signiﬁcant reduction in patients’ trips to healthcare center
(r = 0.891).

Alhassan R et al.
[21]

Out of the 250 adolescents and young adults interviewed, 99% owned
mobile phones. Male young adults were more likely to use mobile
phones for education and prevention of STIs.
Self-eﬃcacy (belief in one’s ability to perform a behavior) was a
signiﬁcant factor.
Amount of telemedicine experience was a factor for future use.
Prior satisfaction with eﬃcacy of telemedicine was a signiﬁcant factor
Positive previous experience was a factor associated with telemedicine
use.
Enjoyment was a factor associated with telemedicine use
Temporal (waiting period), material (travel distance) and social
(judgment of others) dimensions of women’s access to abortion
services intertwined to reduce access and cause discomfort and
inconvenience among women.
Education features to telemedicine enabled decision making
Without telemedicine providing the ﬁrst appointment, this procedure
would not be possible
I prefer telemedicine to avoid my neighbors
Could not go to local clinic out of fear of judgment
Used telemedicine because of privacy (could not go to local clinic from
judgment of others)
Participants indicated beneﬁting from the texts (eg, appreciating
encouraging reminders, coping strategies, and social support) and
suggested improvements (eg, more personalization).
Living below the poverty level predicted worse cessation outcomes at
1-month follow-up among participants receiving in-person only
treatment (P=.03) but not among those receiving iQuit Mindfully.
Reminder calls were more effective than counseling calls or mail alone

Zobair K et al. [22]

Ehrenreich K and
Marston C [22]

Spears CA et al.
[24]

Luckmann R et al.
[26]
Wagstaff A et al.
[26]

Vesnic-Alujevic L
et al. [27]

Tailored education was provided
SMS recipients were more likely to return to clinic in the requested
two days than the control group.

Morita et al. [9]

Moerenhout et al.
[30]

Self-eﬃcacy
User experience using modality
Eﬃcacy of modality
A positive previous experience
Modality is preferred traditional care
Increased convenience (time, miles, work, parking, childcare, etc.)

Care enables other resources (educational/technological)
Increased access
Modality is preferred traditional care
Secure means of care
Increased access
Care enabled through smart mobile technology

Socioeconomics

Eﬃcacy of modality
Care enables other resources (educational/technological)
Care enables other resources (educational/technological)
Reaches developing countries
Increased patient-to-provider communication
Data security

Transmission of data was not secure.
The data were not consistent, had not been scrubbed, were not
veriﬁable

Donelan et al. [20]

Increased convenience (time, miles, work, parking, childcare, etc.)
Care enables other resources (educational/technological)

Preferences for telemedicine were largely due to convenience
Virtual video visits vastly preferred
Follow-up visits through telemedicine were effective
Intervention introduces possibility of shorter, more frequent visits, as
well as earlier intervention.
Clinicians felt a loss of personal connection
Maintained effective communication
Due to the increased cost to the provider, cost would need to be
passed onto the patient. Patients were willing to pay a co-pay
commensurate with the time/cost to travel to the appointment.
Mobile phone applications enable remote care.
Patients reported high satisfaction with the intervention
This modality is effective for certain diseases.
Those without a smartphone may have been less technologically savvy
Evaluates digital self-tracking through health apps and its underlying
paradigm implications
Self-monitoring has greatly increased

Data quality
Increased cost (to provider) because reimbursement does not cover
the technology
Increased convenience (time, miles, work, parking, childcare, etc.)
Modality is preferred traditional care
Eﬃcacy of modality
Improved population health
Less personal
Increased patient-to-provider communication
Increased cost (to provider) because reimbursement does not cover
the technology
Care enabled through smart mobile technology
A positive previous experience
Eﬃcacy of modality
Technical literacy
Improved population health
Self-eﬃcacy
(continued on next page)

222

C.S. Kruse, K. Williams, J. Bohls et al.

Authors

Health Policy and Technology 10 (2021) 209–229

Observations

Observation Themes

Alonso-Solis et al.
This article discusses the implications of health policy for telemedicine
[31]
and how it will enable resources.
Appiah B et al. [32] This article examines how telemedicine can reach developing countries
through mobile phones, which would be the implications of
telemedicine policy.
Bierman et al. [10]

Chabot et al. [4]

CMS [5]
Gonsalves et al.
[33]
Griﬃths et al. [34]

Guo et al. [35]

Katapally [36]
Kjelle et al. [37]

Kotz et al. [1]

Lee et al. [38]

LeGrand et al. [8]

Leung et al. [39]

Melchiorre et al.
[40]

This article discusses what telemedicine health policies can mean for
specialties (OT/PT) and interoperability among states in America. Many
states required specialists to reside within state boundaries to provide
care through telemedicine.
Education provided through mobile devices
Increased access for vulnerable populations
Preferred to sitting in a waiting room
This article examines how health policy can increase healthcare access.
This article discusses how health policies can improve care for certain
diseases, HIV in particular, through telemedicine.
This study examines how health policies can create regulations for
mobile phone applications to increase use of telemedicine.
Increase patient empowerment
Raises issues of ethics
Inadvertent disclosure of information
Raises issues of patient safety

It can improve trust in therapeutic relationships
This article identiﬁed using applications on mobile phones ability to
improve mental health of individuals with HIV.
This article identiﬁes improved care in specialties as an implication of
health policy in telemedicine.
The reimbursement system was not made for this
Without local support, success is diﬃcult
Mobile radiography can reduce transfers of patients to higher levels of
care.
Travel distance and traﬃc was considered
Funding can be a problem
This article discusses how health policies in telemedicine can offer
security.

High and upper middle countries were innovators (early adopters)
Low income countries have fewer resources.
mHealth offers the opportunity to include education and training.
This article discusses implications of a mobile application in regard to
telemedicine for a certain population.
Enables education
Expands services delivered by health professionals
Could be less personal care
Concerns over data security
Very few standards for interfaces
Start-up costs and ogoing maintenance
Certainly, effective for some conditions
Increased communication with provider
Enables self-management
Online schedule is convenient
Observed cost eﬃciencies
Inadequate funding

Saleh et al. [41]

Does not integrates with all EHRs
Lack of skills using eHealth
Privcay concerns
Enables monitoring at a distance
This study identiﬁes implications of telemedicine policies as the ability
to increase resources for those who need to be educated in the subject
and increased access to healthcare services
Users in the intervention group experienced a decrease in HbA1c and
blood pressure
Enables a displaced population access care

Care enables other resources (educational/technological)
Care enabled through smart mobile technology

Reaches developing countries
State licensing

Care enables other resources (educational/technological)
Increased access
Modality is preferred traditional care
Increased access
Improved population health
Improved population health
Self-eﬃcacy
Can raise ethical issues
Data security
Patient safety
Increased cost (to provider) because reimbursement does not cover
the technology
Increased patient-to-provider communication
Care enabled through smart mobile technology
Increased access
Improved population health
Limited reimbursement
Misaligned incentives
Cost advantages
Increased convenience (time, miles, work, parking, childcare, etc.)
Increased cost (to provider) because reimbursement does not cover
the technology
Secure means of care
Data security
Patient privacy
Care enabled through smart mobile technology
Improved population health
Increased access
Improved care for certain diseases
Reaches developing countries
Socioeconomics
Care enables other resources (educational/technological)
Increased patient-to-provider communication
Care enables other resources (educational/technological)
Increased access
Less personal
Data security
Issues of interoperability
Increased cost (to provider) because reimbursement does not cover
the technology
Improved care for certain diseases
Increased patient-to-provider communication
Self-eﬃcacy
Increased convenience (time, miles, work, parking, childcare, etc.)
Cost advantages
Increased cost (to provider) because reimbursement does not cover
the technology
Issues of interoperability
Technical literacy
Patient privacy
Increased access
Care enables other resources (educational/technological)

Increase health outcomes
Increased access
Reaches developing countries
(continued on next page)

223

C.S. Kruse, K. Williams, J. Bohls et al.

Health Policy and Technology 10 (2021) 209–229

Authors

Observations

Observation Themes

Watkins et al. [42]

Patients and health workers used their phones to gather health
information
Users appreciated the convenience of reminders on their mobile
phones
Few could afford the airtime to use devices
Managed chronic diseases
Extends care in rural South Africa
mHealth extends care to cash-strapped elderly patients

Care enables other resources (educational/technological)

Wyatt [43]

Rokicki S and Fink
G [44]

Razmak J and
Belanger C [45]
Grustam et al [46]

Hatef et al [47]
Malvey M and
Slovensky D [3]

Pariyo et al [49]

Dubious safety concerns
Encourages self-care
Eighty-one percent of participants engaged with the mHealth program,
with no evidence that the program disproportionally reached better-off
groups. The program was effective at increasing knowledge of SRH
across all strata. Higher levels of engagement were associated with
higher knowledge scores up to year later. There was no signiﬁcant
impact of the program on self-reported pregnancy within subgroups.
Both groups use the computers and mobile devices regularly for
personal and professional purposes.
Telemedicine can improve the treatment of chronic heart failure
through mHealth
Telemonitoring is safe and reliable
Encourages self-care
Proliferation of cell phone technology make Singapore ideal for
mHealth pilots
Inconsistent reimbursement mechanisms
This article discusses the implication of improving population health
through health policy in telemedicine.
This article addresses several implications of health policy of
telemedicine: improved interoperability, increased access of healthcare,
and dealing with ethical issues.

Sevick et al [50]

mHealth can be highly effective
Service delivery must be equitable
mHealth is ideal for developing countries
Patient appreciated the convenience of the tool.
Readmission avoidance observed
Modality was more effective in 60% of cases
Modality was more costly in 54% of cases

Zelmer et al [11]

Interoperability is a concern

Utidjian L and
Abramson E [51]

Telehealth options prove more convenient around their work schedules
or be more time eﬃcient. It enables care of patients remotely saving
transports that could endanger the patient.
Telehealth a less expensive option because it saves transport costs
between hospitals.
Telehealth as clinically effective as face-to-face
Patient satisfaction is high
Improved access to care
More complete medical record, better continuity of care, closer
follow-up by provider enable more services and higher quality care
with improved safety and less error
Telehealth technology can also provide new ways to educate patients
and their fam-ilies on managing diseases themselves.
Lack of infrastructure in rural areas
Lack of integration into workﬂow
Provider resistance to change
Discomfort with technology
Privacy concerns
Concerns about ability to properly use technology
Loss of face-to-face interaction with providers
Start-up costs and ongoing maintenance
Limited reimbursement opportunities
Unclear ROI
Misaligned incentives (keep patients in-house rather than refer out)
Credentialing at remote sites for referrals
State licensing
Liability protections

Increased convenience (time, miles, work, parking, childcare, etc.)
Socioeconomics
Increase health outcomes
Increased access
Improved care for certain diseases
Increased access
Patient safety
Self-eﬃcacy
Care enables other resources (educational/technological)

Care enables other resources (educational/technological)
Improved care for certain diseases
Increase quality
Self-eﬃcacy
Care enabled through smart mobile technology
Limited reimbursement
Improved population health
Improved interoperability

Increased access
Misaligned incentives
Can raise ethical issues
Improved population health
Socioeconomics
Reaches developing countries
Increased convenience (time, miles, work, parking, childcare, etc.)
Cost advantages
Increase health outcomes
Increased cost (to provider) because reimbursement does not cover
the technology
Issues of interoperability
Increased access
Improved care for certain diseases
Increased convenience (time, miles, work, parking, childcare, etc.)
Increased patient-to-provider communication
Increased convenience (time, miles, work, parking, childcare, etc.)

Cost advantages
Eﬃcacy of modality
A positive previous experience
Increased access
Increase quality

Care enables other resources (educational/technological)
Lack of infrastructure
Integration into workﬂow
Provider resistance
Technical literacy
Patient privacy
Technical literacy
Less personal
Increased cost (to provider) because reimbursement does not cover
the technology
Limited reimbursement
Unclear ROI
Misaligned incentives
Credentialing concerns
State licensing
Liability concerns
(continued on next page)

224

C.S. Kruse, K. Williams, J. Bohls et al.

Health Policy and Technology 10 (2021) 209–229

Authors

Observations

Observation Themes

Martin A et al [52]

Respondents with higher than 10% Medicaid population were more
likely to need consults for orthodontics, whereas those with less than
10% Medicaid were more likely to need periodontal consults
Telehealth as a means of continuing education
Inexpensive modality for arthritis care
Display platforms range from large screen to small mobile devices
Applications are used to communicate experiences and share
information
Research supports a moderately positive effect on clinical outcomes
Central intake system addresses key dimensions of quality
Central intake system uses EMRs with an e-referral system to ensure
all get seen.
eHealth improves self-eﬃcacy
Increased shared decision making
Availability of trained staff
Participants in the intervention group reported higher levels of
satisfaction.
Higher indicators of quality of life (not statistically signiﬁcant)
Higher self-eﬃcacy
Improved anxiety
No national policy in the U.S.
In Estonia, geographical access to hospitals is not considered a
problem.
46 state Medicaid programs reimburse live video, 9 reimburse
store-and-forward, 14 reimburse remote patient monitoring, 3
reimburse all three.
Not all states have passed privacy statutes.
Not all states have passed licensure laws.
Cost is coming down.
This article identiﬁed implications of health policy in telemedicine as
increased access of healthcare, improved quality of healthcare, and the
need to enable resources.

Socioeconomics

Slater H et al [53]

Salisbury C et al
[54]

Rechel B et al [55]

Schmeida M and
McNeal R [56]

Kwong, Mei et al
[57]

Shorten A et al
[58]

Many observations made by the participants about what should and
should not be included in the decision aid.

Knight K et al [59]

Participants described strategies to minimize clinical risk while
brokering competing needs and ﬁnite resources to achieve a “best
outcome” for all stakeholders.

Mukooyo E et al
[60]

Financial management of technology

Care enables other resources (educational/technological)
Cost advantages
Care enabled through smart mobile technology
Care enables other resources (educational/technological)
Increase health outcomes
Increase quality
Increased access
Self-eﬃcacy
Increased patient-to-provider communication
Technical literacy
A positive previous experience
Increase quality
Self-eﬃcacy
Increase health outcomes
State licensing
Increased access
Limited reimbursement

Patient privacy
State licensing
Cost advantages
State licensing

Increased access
Cost advantages
Increase quality
Increase health outcomes
Increased cost (to provider) because reimbursement does not cover
the technology
Patient privacy
Care enables other resources (educational/technological)
Patient privacy
Increased access

A local telephone triage overcomes impersonal nature
Increased cost (to provider) because reimbursement does not cover
the technology
Issues of interoperability
Patient privacy
Increased access

Concerns about standards of interoperability

225

C.S. Kruse, K. Williams, J. Bohls et al.

Health Policy and Technology 10 (2021) 209–229

Appendix C. PICOS

Authors

Sample size

Intervention

Bias

Effect size

Country of origin

Timpel P and Harst Mixed methods
L [28]

32

Survey instrument
on telemedicine

Local population
only

N/A (qualitative)

Germany

Aiyegbajeje F and
Ajayi D [29]

711

Structured
questionnaire on
telemedicine
Structured
questionnaire on
mHealth

Local population
only

N/A

Nigeria

Local population
only

N/A

Ghana

Local population
only

N/A

Australia

Local population
only

N/A

United States

Local population
only

Not reported

United States

Local population
only

Not reported

United States

Local population
only

Medium effect size

EU only

N/A

United States,
Netherlands, and
South Africa
EU

Alhassan R et al
[21]

Zobair K et al [22]

Ehrenreich K and
Marston C [22]
Spears CA et al
[24]

Luckmann R et al
[26]

Study design

Participants

Hospital patients
participating in
telemedicine
services
Structured
Patients at Lagos
questionnaire
University
Teaching Hospital
Cross-sectional
Students at one
analytical study
university in
through structured Ghana
questionnaire
Cross-sectional
Rural citizens of
survey
Bangladesh

250

492

Qualitative analysis Customers of
20
Planned
Parenthood of Utah
RCT
Smoking cessation 71
patients in
low-income area of
Atlanta, GA
30162
RCT
primary care
population with a
high baseline
mammography
adherence

Structured
questionnaire on
telemedicine
Structured
interviews on
telemedicine
mHealth

Phone calls to
educate and
remind patients
about the
importance of
regular
mammography
SMS reminders
that treatment of
TB is necessary
Wearable sensors

Wagstaff A et al
[26]

RCT

TB patients at Cape 506
Town clinics

Vesnic-Alujevic L
et al [27]
Donelan et al [20]

Opinion piece

N/A

N/A

Cross-sectional
survey

315

Virtual video visits Convenience
sample

Morita et al [9]

Cross-sectional
experiemt

254 patients & 61
clinicians
participating in
telehealth
138 asthma
patients

138

Moerenhout et al
[30]
Alonso-Solis et al
[31]

Opinion piece

N/A

mHealth platform
on phones and
tablets for asthma
N/A

Surveillance effect, Effect size not
selection bias
mentioned
(high % female)
none
N/A

Prospective
feasibility study
protocol

45
Schizophrenic
patients in Israel,
Hungary, and Spain

mHealth platform
for
self-management
of schitzophrenia.
Providers would be
a psychiatrist,
psychologist, and a
case manager.
Caller tunes as
incentive to
change behavior
and encourage
blood donation
Telemedicine for
occupational
therapy and
physical therapy.
STBBI testing
service

none

No results
(protocol)

Spain, Hungary,
Israel

Local population
only

Effect size not
mentioned

Ghana

US only

Effect size not
mentioned

United States

Local population
only

N/A

Canada

N/A

Appiah B et al [32] Cross-sectional,
true experiment

Subsaharan Africa

955

Bierman et al [10]

Surveyed 50 states

All 50 states

50

Chabot et al [4]

Structured
interviews

37
Healthcare
providers,
administrators,
policy makers, and
community-based
service providers
involved with
STBBI testing
service

Effect size not
mentioned

United States

Canada

Belgium

Qualitative analysis
(continued on next page)
226

C.S. Kruse, K. Williams, J. Bohls et al.

Authors

Study design

CMS [5]

White paper

Gonsalves et al
[33]
Griﬃths et al [34]

Health Policy and Technology 10 (2021) 209–229

Participants

Sample size

None –
0
announcement of
new
reimbursement
mechanisms in
2019
Bayesian algorithm Bayesian algorithm 250

Intervention

Bias

Effect size

Country of origin

All types of
telemedicine

US only

N/A

United States

US only

N/A

United States

Local population
only
China only

N/A

UK

No results
(protocol)

China

Effect size not
mentioned
N/A

Canada
Norway

N/A

United States

N/A

Sub-Saharan Africa

Guo et al [35]

RCT

Katapally [36]

Longitudinal

Canadians

317

Computer model
only
Digital
communications
Smart-phone app
aimed at mental
health of PLWH
Smart-phone app

Kjelle et al [37]

Semi-structured
interviews
Delphi study

Norway

11

Mobile radiography Norway only

0

Wearable and
No results
portable
(protocol)
monitoring devices
mHealth
Local population
only

Kotz et al [1]

Lee et al [38]

LeGrand et al [8]

Leung et al [39]

Melchiorre et al
[40]
Saleh et al [41]
Watkins et al [42]
Wyatt [43]
Rokicki S and Fink
G [44]
Razmak J and
Belanger C [45]
Grustam et al [46]

Mixed methods

Younger patients
of England’s NHS
People living with
HIV

None – discussion
of wearable
monitoring devices
Survey from WHO Low and middle
income countries
surveyed from
WHO
Two-group,
Young men who
active-control RCT have sex with men
and transgender
women who have
sex with men
Exploratory study, Chinese citizens.
qualitative analysis Message threads
were examined.
Qualitative survey health programs in
analysis
Europe
RCT
Diabetic patients
in Lebanon
Semi-structured
Patients and
interviews
providers
Opinion piece
None
Experimental
Girls aged 14-22 in
22 secondary
schools
Structured
Physicians and
patients in
interviews
Northern Ontario
Computer model
Chronic heart
failure patients

433
300

112

Smart-phone app

No results
(protocol)

No results
(protocol)

United States

571

Social media
platform

China only

N/A

China

101

eHealth

Europe only

N/A

EU

2359

mHealth
mHealth

Effect size not
mentioned
N/A

Lebanon

156

South Africa

0
498

mHealth
mHealth

Refugee camps in
Lebanon only
Rural South Africa
only
none
Ghana only

N/A
N/A

UK
Ghana

323 physicians and eHealth
325 patients

Canada only

N/A

Canada

0

none

N/A

441

mHealth

Singapore,
Netherlands, and
United States only
Baltimore only

Singapore,
Netherlands,
United States
United States

none

none

N/A

United States

Low and middle
income countries
only
Canada only

N/A

United States

Effect size not
mentioned

Canada

Local population
only

Effect size not
mentioned

Canada

Hatef et al [47]

Cross sectional
study

Malvey M and
Slovensky D [3]
Pariyo et al [49]

Opinion piece

Diabetic patients
in the East
Baltimore Medical
Center
none

Opinion piece

none

none

mHealth

Sevick et al [50]

Cross-sectional
experiment

Patients in Canada

1354

Zelmer et al [11]

Secondary data
analysis

Countries using
information and
communication
technology in care
N/A

38

mHealth,
electronic
discharge
communication
tool
mHealth,
telemedicine

Utidjian L and
White paper
Abramson E [51]
Martin A et al [52] Qualitative analysis South Carolina
Dental Association
members

Canada only

Effect size not
mentioned

N/A

N/A

US only

N/A

United States

384

Telehealth for
dentistry

US only

N/A

United States

(continued on next page)

227

C.S. Kruse, K. Williams, J. Bohls et al.

Health Policy and Technology 10 (2021) 209–229

Authors

Study design

Participants

Sample size

Intervention

Bias

Effect size

Country of origin

Slater H et al [53]

White paper

N/A

N/A

eHealth

N/A

Australia

Salisbury C et al
[54]

RCT

609

eHealth

N/A

UK

Rechel B et al [55]

Delphi study

patients of 43
general practices
in England
Key experts in 8
countries

Local population
only
Local population
only

Not reported

Telehealth

None detected

N/A

Schmeida M and
McNeal R [56]
Kwong, Mei et al
[57]
Shorten A et al
[58]

Multivariate
regression analysis
White paper

N/A

N/A

N/A

US only

N/A

Europe, United
States, Canada,
Australia, Italy,
Spain, UK, Croatia,
Estonia
United States

N/A

N/A

N/A

none

N/A

United States

Knight K et al [59]
Mukooyo E et al
[60]

Iterative and
Pregnant women
participatory
at one hospital
research approach
Qualitative analysis Registered nurses

9

eHealth/mHealth
decision aid

US only

N/A

United States

15

Telephone triage

N/A

Australia

Delphi study

163

eHealth

Local population
only
Local population
only

N/A

Uganda

Those who had
particpated in
eHealth activities

[21] Alhassan RK, Abdul-Fatawu A, Adzimah-Yeboah B, Nyaledzigbor W, Agana S,
Mwini-Nyaledzigbor P. Determinants of use of mobile phones for sexually transmitted infections (STIs) education and prevention among adolescents and young adult population in Ghana: implications of public health
policy and interventions design. Reprod Health 2019;16(1):120. doi:10.1186/
s12978- 019- 0763- 0.
[22] Zobair KM, Sanzogni L, Sandhu K. Expectations of telemedicine health service
adoption in rural Bangladesh. Soc Sci Med 2019:238. doi:10.1016/j.socscimed.
2019.112485.
[23] Ehrenreich K, Marston C. Spatial dimensions of telemedicine and abortion access: a qualitative study of women’s experiences. Reprod Health 2019;16(1):94.
doi:10.1186/s12978- 019- 0759- 9.
[24] Spears CA, Abroms LC, Glass CR, et al. Mindfulness-based smoking cessation
enhanced with mobile technology (iQuit mindfully): pilot randomized controlled trial. JMIR mHealth uHealth 2019;7(6). doi:10.2196/13059.
[25] Luckmann R, Costanza ME, White MJ, et al. A 4-year randomized trial comparing three outreach interventions to promote screening mammograms. Transl
Behav Med 2019;9(2):328–35. doi:10.1093/tbm/iby031.
[26] Wagstaff A, van Doorslaer E, Burger R. SMS nudges as a tool to reduce tuberculosis treatment delay and pretreatment loss to follow-up. A randomized
controlled trial. PLoS ONE 2019;14(6):1–14. doi:10.1371/journal.pone.0218527.
[27] Vesnic-Alujevic L, Pereira AG, Breitegger M. Wearable sensors. US: Academic
Press; 2019.
[28] Timpel P, Harst L. Research implications for future telemedicine studies and
innovations in diabetes and hypertension—a mixed methods study. Nutrients
2020;12(5). doi:10.3390/nu12051340.
[29] Aiyegbajeje FO, Ajayi DD. Mobile health communication: effective reduction
of patients’ travel needs to hospitals. Int J Healthc Manage 2020;13(1):28–34.
doi:10.1080/20479700.2018.1434862.
[30] Moerenhout T, Devisch I, Cornelis GC. E-health beyond technology: analyzing
the paradigm shift that lies beneath. Med Health Care Philos 2018;21(1):31–
41. doi:10.1007/s11019-017-9780-3.
[31] Alonso-Solis A, Rubinstein K, Corripio I, et al. Mobile therapeutic attention for
treatment-resistant schizophrenia (m-RESIST): a prospective multicentre feasibility study protocol in patients and their caregivers. BMJ Open 2018;8(7).
doi:10.1136/bmjopen- 2017- 021346.
[32] Appiah B, Burdine JN, Aftab A, et al. Determinants of intention to use mobile
phone caller tunes to promote voluntary blood donation: cross-sectional study.
JMIR mHealth uHealth 2018;6(5). doi:10.2196/mhealth.9752.
[33] Gonsalves GS, Copple JT, Johnson T, Paltiel AD, Warren JL. Bayesian adaptive
algorithms for locating HIV mobile testing services. BMC Med 2018;16(1):155.
doi:10.1186/s12916-018-1129-0.
[34] Griﬃths F, Bryce C, Cave J, et al. Timely digital patient-clinician communication
in specialist clinical services for young people: a mixed-methods study (the
LYNC study). J Med Internet Res 2017;19(4):e102. doi:10.2196/jmir.7154.
[35] Guo Y, Alicia Hong Y, Qiao J, et al. Run4Love, a mHealth (WeChat-based)
intervention to improve mental health of people living with HIV: a randomized controlled trial protocol. BMC Public Health 2018(1):1. doi:10.1186/
s12889- 018- 5693- 1.
[36] Katapally TR, Bhawra J, Leatherdale ST, et al. The smart study, a mobile health
and citizen science methodological platform for active living surveillance, integrated knowledge translation, and policy interventions: longitudinal study. J
Med Internet Res 2018;20(3). doi:10.2196/publichealth.8953.
[37] Kjelle E, Kristin BL, Hilde MO, Aud MM. Managers’ experience of success
criteria and barriers to implementing mobile radiography services in nurs-

References
[1] Kotz D, Lord SE, O’Malley AJ, Stark L, Marsch LA. Workshop on emerging technology and data analytics for behavioral health. J Med Internet Res 2018;20(6).
doi:10.2196/resprot.9589.
[2] Foley M. HIT in 2017 and for the long haul. Health Manage Technol
2017;38(2):6–13.
[3] Malvey DM, Slovensky DJ. Global mHealth policy arena: status check and future directions. Mhealth 2017;3:41. doi:10.21037/mhealth.2017.09.03.
[4] Chabot, C., Gilbert, M., Haag, D., et al. Anticipating the potential for positive uptake and adaptation in the implementation of a publicly funded online
STBBI testing service: a qualitative analysis. BMC Health Serv Res 2018(1):1.
doi: 10.1186/s12913-018-2871-x.
[5] CMS adds new modiﬁers, expands telehealth policies for 2019 in physician
rule. BRIEF APCS 2018;19(12):7–9.
[6] (OCR), Oﬃce for Civil Rights. HITECH act enforcement interim ﬁnal rule.
HHS.gov Web site. Updated 2009. Accessed Sep 21, 2019.
[7] World Health Organization. Telemedicine: opportunities and developments in
member states. Report on the second global survey on eHealth. World Health
Organization. 2010.
[8] LeGrand S, Muessig KE, Platt A, et al. Epic allies, a gamiﬁed mobile phone app
to improve engagement in care, antiretroviral uptake, and adherence among
young men who have sex with men and young transgender women who have
sex with men: protocol for a randomized controlled trial. JMIR Res Protoc
2018;7(4):e94. doi:10.2196/resprot.8811.
[9] Morita PP, Yeung MS, Ferrone M, et al. A patient-centered mobile health system that supports asthma self-management (breathe): design, development,
and utilization. JMIR mHealth uHealth 2019;7(1):e10956. doi:10.2196/10956.
[10] Bierman R, Trey KMW, Calouro C. State occupational and physical therapy telehealth laws and regulations: a 50-state survey. Int J Telerehabil 2018(2):3.
doi:10.5195/ijt.2018.6269.
[11] Zelmer J, Ronchi E, Hyppönen H. et al. International health IT benchmarking: learning from cross-country comparisons. J Am Med Inform Assoc
2017;24(2):371–9. doi:10.1093/jamia/ocw111.
[12] Kruse CS. Writing a systematic review for publication in a health-related degree program. JMIR Res Protoc 2019;8(10):e15490.
[13] . Telemedicine. World Health Organization; 2010. p. 2.
[14] Thomas BH, Ciliska D, Dobbins M, Micucci S. A process for systematically reviewing the literature: providing the research evidence for public health nursing interventions. Worldviews Evid Based Nurs 2004;1(3):176–84.
[15] Violet P. Johns Hopkins nursing evidence-based practice: model and guidelines. 2nd edition. AORN J 2013(1):157. doi:10.1016/j.aorn.2012.10.004.
[16] Saletta M, Kruger A, Primoratz T, Barnett A, van Gelder T, Horn RE. The role
of narrative in collaborative reasoning and intelligence analysis: a case study.
PLoS ONE 2020;15(1):1–17. doi:10.1371/journal.pone.0226981.
[17] Braun V, Clarke V. Using thematic analysis in psychology. Qual Res Psychol
2006;3(2):77–101. doi:10.1191/1478088706qp063oa.
[18] McHugh ML. Interrater reliability: the Kappa statistic. Biochem Med
2012;22(3):276–82.
[19] Light RJ. Measures of response agreement for qualitative data: some generalizations and alternatives. Psychol Bull 1971;76(5):365.
[20] Donelan K, Barreto EA, Sossong S, et al. Patient and clinician experiences with
telehealth for patient follow-up care. Am J Manage Care 2019;25(1):40–4.
228

C.S. Kruse, K. Williams, J. Bohls et al.

Health Policy and Technology 10 (2021) 209–229

ing homes in Norway: a qualitative study. BMC Health Serv Res 2018(1):1.
doi:10.1186/s12913-018-3115-9.
[38] Lee S, Begley CE, Morgan R, Chan W Chan, Kim S-Y. m-Health policy readiness and enabling factors: comparisons of Sub-Saharan Africa and organization for economic cooperation and development countries. Telemed J E Health
2018;24(11):908–21. doi:10.1089/tmj.2017.0278.
[39] Leung R, Guo H, Pan X. Social media users’ perception of telemedicine and
mHealth in China: exploratory study. JMIR mHealth uHealth 2018;6(9). doi:10.
2196/mhealth.7623.
[40] Melchiorre MG, Papa R, Rijken M, van Ginneken E, Hujala A, Barbabella F.
eHealth in integrated care programs for people with multimorbidity in Europe: insights from the ICARE4EU project. Health Policy 2018;122(1):53–63.
doi:10.1016/j.healthpol.2017.08.006.
[41] Saleh S, Farah A, Dimassi H, et al. Using mobile health to enhance outcomes of
noncommunicable diseases care in rural settings and refugee camps: randomized controlled trial. JMIR mHealth uHealth 2018;6(7). doi:10.2196/mhealth.
8146.
[42] Watkins JO, Goudge J, Gómez-Olivé FX, Griﬃths F. Mobile phone use among
patients and health workers to enhance primary healthcare: a qualitative study
in rural South Africa. Soc Sci Med 2018;198:139–47. doi:10.1016/j.socscimed.
2018.01.011.
[43] Wyatt JC. Correction to: how can clinicians, specialty societies and others evaluate and improve the quality of apps for patient use? BMC Med 2019;17(1).
doi:10.1186/s12916- 019- 1381- y.
[44] Rokicki S, Fink G. Assessing the reach and effectiveness of mHealth: evidence
from a reproductive health program for adolescent girls in Ghana. BMC Public
Health 2017;17(1):1–14. doi:10.1186/s12889-017-4939-7.
[45] Razmak J, Bélanger CH. Comparing Canadian physicians and patients on their
use of e-health tools. Technol Soc 2017;51:102–12. doi:10.1016/j.techsoc.2017.
08.007.
[46] Grustam AS, Vrijhoef HJM, Koymans R, Hukal P, Severens JL. Assessment of
a business-to-consumer (B2C) model for telemonitoring patients with chronic
heart failure (CHF). BMC Med Inform Decis Mak 2017;17(1):145. doi:10.1186/
s12911- 017- 0541- 2.
[47] Hatef E, Alexander M, Vanderver BG, Fagan P, Albert M. Assessment of annual
diabetic eye examination using telemedicine technology among underserved
patients in primary care setting. Middle East Afr J Ophthalmol 2017;24(4):207–
12. doi:10.4103/meajo.MEAJO_19_16.
[48] Opoku D, Stephani V, Quentin W. A realist review of mobile phone-based
health interventions for non-communicable disease management in SubSaharan Africa. BMC Med 2017;15(1):24. doi:10.1186/s12916- 017- 0782- z.
[49] Pariyo GW, Wosu AC, Gibson DG, Labrique AB, Ali J, Hyder AA. Moving
the agenda on noncommunicable diseases: policy implications of mobile
phone surveys in low and middle-income countries. J Med Internet Res
2017;19(5):e115. doi:10.2196/jmir.7302.

[50] Sevick LK, Santana M, Ghali WA, Clement F. Prospective economic evaluation
of an electronic discharge communication tool: analysis of a randomised controlled trial. BMJ Open 2017;7(12). doi:10.1136/bmjopen- 2017- 019139.
[51] Utidjian L, Abramson E. Pediatric telehealth: opportunities and challenges. Pediatr Clin North Am 2016;63(2):367–78. https://doi.org/10.1016/j.pcl.2015.11.
006.
[52] Martin AB, Nelson JD, Bhavsar GP, McElligott J, Garr D, Leite RS. Feasibility assessment for using telehealth technology to improve access to dental care for
rural and underserved populations. J Evid Based Dent Pract 2016;16(4):228–35.
https://doi.org/10.1016/j.jebdp.2016.08.002.
[53] Slater H, Dear BF, Merolli MA, Li LC, Briggs AM. Use of eHealth technologies to
enable the implementation of musculoskeletal models of care: evidence and
practice. Best Prac Res Clini Rheumatol 2016;30(3):483–502. https://doi.org/10.
1016/j.berh.2016.08.006.
[54] Salisbury C, O’Cathain A, Edwards L, et al. Effectiveness of an integrated telehealth service for patients with depression: a pragmatic randomised controlled trial of a complex intervention. Lancet Psychiatry 2016;3(6):515–25.
https://doi.org/10.1016/S2215-0366(16)0 0 083-3.
[55] Rechel B, Džakula A, Duran A, et al. Hospitals in rural or remote areas:
an exploratory review of policies in 8 high-income countries. Health Policy
2016;120(7):758–69. https://doi.org/10.1016/j.healthpol.2016.05.011.
[56] Schmeida M, McNeal R. State policy action on medicaid telemedicine reimbursement laws. Health Policy Technol 2016;5(1):32–9. https://doi.org/10.1016/
j.hlpt.2015.10.007.
[57] Kwong MW, Calouro C, Nasseri L, Gutierrez M. The federally funded telehealth
resource centers. Perspect Telepract 2015;5(1):14–23. doi:10.1044/tele5.1.14.
[58] Shorten A, Fagerlin A, Illuzzi J, Kennedy HP, et al. Developing an internetbased decision aid for women choosing between vaginal birth after cesarean
and planned repeat cesarean. J Midwifery Women’s Health 2015;60(4):390–
400. doi:10.1111/jmwh.12298.
[59] Knight, K., Kenny, A., Endacott, R.. Assessing clinical urgency via telephone in
rural Australia. Nurs Health Sci. 2015(2):201. doi: 10.1111/nhs.12161.
[60] Mukooyo ES, Lutwama A, Munabi IG, Sewankambo N, Aceng RJ. A mixed methods approach to prioritizing components of Uganda’s eHealth environment.
Pan Afr Med J 2015:22. doi:10.11604/pamj.2015.22.198.7192.
[61] Browne FA, Wechsberg WM, Kizakevich PN, et al. mHealth versus face-to-face:
study protocol for a randomized trial to test a gender-focused intervention for
young African American women at risk for HIV in North Carolina. BMC Public
Health 2018;18(1). doi:10.1186/s12889-018-5796-8.
[62] Hollander JE, Carr BG. Virtually perfect? Telemedicine for COVID-19. N Engl J
Med 2020. doi:10.1056/NEJMp2003539.
[63] Dube JF. Fraud in health care and organized crime. Med Health R I
2011;94(9):268–9.
[64] Svorny S. Telemedicine… across state borders. Milken Institute Review Web
site 2019 Accessed Sep 21.

229

"""
short_summary = summarize(original_text)
print(short_summary)

summary_by_word_count=summarize(original_text,word_count=150)
print(summary_by_word_count)
```
